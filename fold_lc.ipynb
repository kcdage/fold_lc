```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##creating a fake lightcurve starting from some initial observation time t_init

N=100
t_init=59754
t=np.arange(0,N, 2)+t_init


##generating periodic data 
period=3.412
clean_data=7*np.sin(2*np.pi*t/period)+10



##adding in noise
noise_ts = 5* (np.random.rand(len(t)) - 0.5)
data=clean_data+noise_ts
```

```python
plt.plot(t,data)
plt.xlabel('Time (arbitrary units)')
plt.ylabel('Flux (arbitrary units)')
```

```python
##folding the lightcurve using the remainder function, and subtracting the initial time of observation

phase=((t-t_init) % period)/period
```

```python
plt.scatter(phase,data)
plt.xlabel('phase')
plt.ylabel('flux')
```
